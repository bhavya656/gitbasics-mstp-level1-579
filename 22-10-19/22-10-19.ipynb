```python
# draw a plot for time and distance by using list comphrension
import matplotlib.pyplot as pl
distance=list(range(1,10,3))
time=[3,5,6,7,8]
distance=2*time
pl.hist(distance,time)
pl.xlabel('dis=Data')
pl.ylabel('time-Axis')
pl.title('time and distance')
pl.grid()
pl.show()
```

# Representing a sine wave
A=sine(wt)
A=Amplitude
w=frequency
t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title("sine wave")
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```
