# creating subplot using numpy with sine

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s)
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")

```

# assiging all null values

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="1st graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2nd graph with sign wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()





```

# hist,scatter using subplot

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(222)
plt.plot(np.random.rand(30))
ax1.hist(np.random.rand(50),bins=20,color='m')
ax2.scatter(np.arange(30),np.arange(30)*8)
```

# 3d graph

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import style
style.available
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
plt.show()
```

```python
x=[3,4,2,1]
y=[5,4,99,8]
z=[2,5,4,3]
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z)
plt.show()
plt.xlabel("x")
plt.ylabel("y")
plt.title("3d projection")
plt.show()
```

# bar

```python
x=[3,4,2,1]
y=[5,4,99,8]
z=[2,5,4,3]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.bar3d(x,y,z,dx,dy,dz,color='m')
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel('z')
plt.title("3d projection")
plt.show()
```

# scatter

```python
x=[3,4,2,1]
y=[5,4,99,8]
z=[2,5,4,3]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.scatter(x,y,z,color='k')
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel('z')
plt.title("3d projection")
plt.show()
```
